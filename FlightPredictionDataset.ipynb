```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
frames = [df_train, df_test]
final_df = pd.concat(frames)
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.isnull().sum()
```

```python
### Converting the Date of Journey Column to datatime datatype

final_df['Date_of_Journey'] = pd.to_datetime(final_df['Date_of_Journey'])
```

```python
final_df.head()
```

```python
### Separating the various aspect of the date into respective day,month, and year

final_df['Day'] = final_df['Date_of_Journey'].dt.day
final_df['Month'] = final_df['Date_of_Journey'].dt.month
final_df['Year'] = final_df['Date_of_Journey'].dt.year
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop(['Date_of_Journey'], axis = 1, inplace=True)
```

```python
final_df.head()
```

```python
### Converting the Arrival time to date datatype and to hour and minutes and converting the results into intergers

final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute'] = final_df['Arrival_Minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(['Arrival_Time'], axis = 1, inplace = True)
```

```python
final_df.tail(1)
```

```python
### Converting the Depature time to date datatype and to hour and minutes and converting the results into intergers

final_df['Dep_Hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minute'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Dep_Hour']=final_df['Dep_Hour'].astype(int)
final_df['Dep_Minute']=final_df['Dep_Minute'].astype(int)
```

```python
final_df.head(1)
```

```python
final_df.drop(['Dep_Time'], axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
final_df.info()
```
